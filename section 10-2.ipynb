```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sns.get_dataset_names()
```

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

```python
grouped=df.groupby('key1')
grouped.describe().transpose()
```

```python
grouped['data1'].quantile(0.9)
```

```python
# print(grouped.max())
# grouped.min()
```

```python
def data_limit(arr):
    return arr.max() - arr.min()

grouped.agg(data_limit)
```

```python
grouped.describe().transpose()
```

# Column-Wise and Multiple Function Application

```python
# tips = pd.read_csv('examples/tips.csv')
# # Add tip percentage of total bill

# tips['tip_pct'] = tips['tip'] / tips['total_bill']
# tips[:6]
```

```python
# another way
tips = sns.load_dataset('tips')
tips.info()
```

```python
tips.head(4)
```

```python
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] )*100
tips[:6]
```

```python
grouped = tips.groupby(['day', 'smoker'])
grouped.describe().transpose().head(11)
```

```python
grouped_pct = grouped['tip_pct']
# grouped_pct.agg('mean')

# or
grouped['tip_pct'].mean()
```

```python
grouped_pct.agg(['mean', 'std', data_limit])
```

```python
grouped_pct.agg([('average', 'mean'), ('std. dev.',"std")])
```

```python
functions = ['count', 'mean', 'max']
result = grouped['tip_pct', 'total_bill'].agg(functions)
result
```

```python
result['tip_pct']
```

```python
ftuples = [('AVERAGE', 'mean'), ('VARIENCE', np.var)]
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
grouped.agg({'tip' : np.max, 'size' : 'sum'})
# print(tips.head())
```

```python
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],
             'size' : 'sum'})
```

```python
# or 
grouped.agg({'tip_pct' : ['min', 'max', ('average','mean'), 'std'],
             'size' : 'sum'})
```

# Returning Aggregated Data Without Row Indexes

```python
tips.groupby(['day', 'smoker'], as_index=False).mean()
```
